## Import Packages


```python
import pandas as pd
from scipy.stats import chi2_contingency, ttest_ind
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
import matplotlib.pyplot as plt
```

## Chi-Square Test

```python
COV = pd.read_csv('https://raw.githubusercontent.com/Stephen-sanchez1/datasci_5_statistics/main/data/COVID-19_Cases_and_Deaths_by_Age_Group_-_ARCHIVE.csv')
COV
```

```python
COV['AgeGroups'].value_counts()
COV['Confirmed cases'].value_counts()

contingency_table = pd.crosstab(COV['AgeGroups'], COV['Confirmed cases'])
print(contingency_table)

chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Chi2 value: {chi2}")
print(f"P-value: {p}")
```
